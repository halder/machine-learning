```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (15, 10)
```

```python
N = 50
X = np.linspace(0, 10, N)
Y = 0.5*X + np.random.randn(N)
Y[-1] += 30
Y[-2] += 30
```

```python
plt.scatter(X, Y)
```

```python
X = np.vstack([np.ones(N), X]).T
```

```python
def r2(x, y):
    w = np.linalg.solve(np.dot(x.T, x), np.dot(x.T, y))
    yhat = np.dot(x, w)
    
    dp = y - yhat
    dm = y - y.mean()
    
    # L2 regularization:
    
    # lambda, hyperparameter
    l2 = 1000
    w_map = np.linalg.solve(l2*np.eye(2) + np.dot(X.T, X), np.dot(X.T, Y))
    yhat_m = np.dot(x, w_map)
    
    dpm = y - yhat_m
    dmm = y - y.mean()
    
    
    plt.scatter(x[:,1], y)
    plt.plot(x[:,1], yhat, c="orange")
    plt.plot(x[:,1], yhat_m, c="green")
    
    return 1 - ( np.dot(dp, dp) / np.dot(dm, dm) )
```

```python
r2(X, Y)
```
