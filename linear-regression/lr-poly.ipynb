```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (15, 10)
```

```python
file = "data_poly.csv"
```

To use multiple linear regression on a polynomial, the input table (X) needs to be created differently.\
Instead of only appending x to X, x*x (the polynomial term) needs to be added to the matrix X aswell. Everything else is the same process as for the 2d linear regression.

```python
X, Y = [], []
for line in open(file):
    x, y = line.split(",")
    x = float(x)
    # add x AND x*x to big X
    X.append([x, x*x, 1]), Y.append(float(y))
X, Y = np.array(X), np.array(Y)
```

```python
# visualize data
plt.scatter(X[:,0], Y)
```

```python
w = np.linalg.solve(np.dot(X.T, X),np.dot(X.T, Y))
```

```python
Yhat = np.dot(X, w)
```

```python
d_pred = Y - Yhat
d_mean = Y - Y.mean()
```

```python
r2 = 1 - (d_pred.dot(d_pred) / d_mean.dot(d_mean))
r2
```
